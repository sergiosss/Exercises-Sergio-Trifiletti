```python
%pylab inline
from scipy.special import binom
import numpy as np
import numpy.polynomial.chebyshev as cheb
```

1) Make a function that computes the L^infty norm of the error for the interpolation of a given function f when using polynomial approximation of increasing order with the following three basis:

        Equispaced Lagrange interpolation
        Chebishev Lagrange interpolation
        Bernstein approximation



```python
def standard_B(i,n,x): #Returns the i-th Bernstein polynomial of degree n
    # evaluated in the point x with the standard definition.
    return binom(n,i)*(x**i)*(1-x)**(n-i)

def B_approx(n,f,x): #Returns the Bernstein approximation of degree n
    # of the function f evaluated in the point x.
    sum = 0
    for i in range(n+1):
        sum += standard_B(i,n,x)*f(i*1.0/n)
    return sum
    
def l(q,i,x): #The i-th Lagrange polynomial on the interpolation points q
    n = len(q)
    prod = 1
    for j in range(n):
        if j != i:
            prod *= (x - q[j]) / (q[i] - q[j])
    return prod

def L(q,g,x): #The Lagrange interpolation polynomial of the function g
    # on the interpolation points q
    n = len(q)
    sum = 0
    for i in range(n):
        sum += l(q,i,x) * g(q[i])
    return sum

def Runge(x):
    return 1.0/(1.0 + (5.0-10.0*x)**2)

def L_infty(f): #Returns the L-infty norm of the function f on the
    # interval [0,1]. It evaluates the function in an increasing number
    # of points (starting from 1025) and takes the maximum.
    # It ends when the results of two iterations differ by less than the
    # tollerance value tol or when the number of evaluation points is too
    # big (in this case it shows a warning message).
    m = 2**10
    tol = 1e-3
    maxlen = 2**24
    x = np.linspace(0,1,m+1)
    y = abs(f(x))
    new = max(y)
    old = new - 1
    while m<maxlen and abs(new-old)>tol:
        m *= 2
        old = new
        x = np.linspace(0,1,m+1)
        y = abs(f(x))
        new = max(y)
    if abs(new-old)>tol:
        print("WARNING: Inaccurate result")
    return new
```

```python
def error(fun,meth,n): #Returns the L-infty norm of the difference
    # between the function with index fun (0<=fun<=2) and its approximation
    # of order n with the method meth (0<=meth<=2).
    f = []
    f.append(lambda x: np.sin(2.0*np.pi*x))
    f.append(lambda x: Runge(x))
    f.append(lambda x: abs(x-0.5))
    if meth==0:
        q = np.linspace(0,1,n)
        dif = lambda x: f[fun](x)-L(q,f[fun],x)
        return L_infty(dif)
    elif meth==1:
        q = cheb.chebgauss(n)[0][::-1]*.5+.5
        dif = lambda x: f[fun](x)-L(q,f[fun],x)
        return L_infty(dif)
    elif meth==2:
        dif = lambda x: f[fun](x)-B_approx(int(n),f[fun],x)
        return L_infty(dif)
    else:
        return None
```

2) Plot the error VS number_of_interpolation_points for the following functions:

        f(x) = sin(2*pi*x)
        Runge function
        f(x) = abs(x-.5)



```python
meths = ['Lagrange equispaced', 'Lagrange Chebyshev', 'Bernstein']
funcs = ['sin(2\pi x)','1/(1+(5-10x)^2)','|x|']
Ns = range(10,51)

for i in range(3):
    for j in range(3):
        figure()
        plot(Ns,[error(i,j,n) for n in Ns])
        xlabel('order of approximation')
        ylabel('error')
        title(meths[j] + ', $f(x)=' + funcs[i] + '$')
```

Lagrange interpolation with equispaced interpolation points behaves terribly as the number of points increases, while Lagrange interpolation with Chebyshev points commits an error which tends to $0$. Bernstein polynomial's error tends to $0$ too, but in a slower way.

In the following graphs are compared, for each function, the errors committed with each method.

```python
line = '234'
for i in range(3):
    figure()
    for j in range(3):
        plot(Ns,[error(i,j,n) for n in Ns],line[j],label=meths[j])
    legend()
    xlabel('order of approximation')
    ylabel('error')
    title('$f(x)=' + funcs[i] + '$')
```

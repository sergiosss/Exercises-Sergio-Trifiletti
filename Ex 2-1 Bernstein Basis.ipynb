1) Construct a function that implements the ith Bernstein basis, using the definitions in http://www.idav.ucdavis.edu/education/CAGDNotes/Bernstein-Polynomials.pdf

        The standard definition on top of page 2
        The recursive definition on top of page 5
        The power basis definition on bottom of page 8

```python
def standard_B(i,n,x): #Returns the i-th Bernstein polynomial of degree n
    # evaluated in the point x with the standard definition.
    return binom(n,i)*(x**i)*(1-x)**(n-i)
```

```python
def recurs_B(i,n,x): #Returns the i-th Bernstein polynomial of degree n
    # evaluated in the point x with the recursive definition.
    if i==0:
        return (1-x)**n
    elif i==n:
        return x**n
    else:
        return (1-x)*recurs_B(i,n-1,x)+x*recurs_B(i-1,n-1,x)
```

```python
def power_B(i,n,x): #Returns the i-th Bernstein polynomial of degree n
    # evaluated in the point x with the power basis definition.
    y = 0
    for j in range(i,n+1):
        y += ((-1.0)**(j-i))*binom(n,j)*binom(j,i)*(x**i)
    return y
```

2) Compare running times for the three options above for n=20, using the builtin cProfile module

```python
n = 20
cProfile.run('for i in range(n+1): standard_B(i,n,0.5)')
```

```python
cProfile.run('for i in range(n+1): recurs_B(i,n,0.5)')
```

```python
cProfile.run('for i in range(n+1): power_B(i,n,0.5)')
```

It is evident that the recursive definition takes a huge amount of operations (hence time). Power basis definition, even if a bit slower than the standard one, is quite close to it.

3) Plot all basis obtained through options a) and c) for n=20,40,80,160,320. Give an explanation for what you observe.

```python
x = np.linspace(0,1,1025)
Ns = [20, 40, 80, 160, 320]
for j in range(len(Ns)):
    figure()
    n = Ns[j]
    subplot(1, 2, 1)
    for i in range(n+1):
        plot(x, standard_B(i,n,x))
    title('standard')
    subplot(1, 2, 2)
    for i in range(n+1):
        plot(x, power_B(i,n,x))
    title('power')
```

The power basis definition is dramatically inaccurate, which is what we expected since it involves a lot of additions (ill-conditioned operation).

4) Construct a function that returns the derivative of the ith Bernstein basis, evaluated at x.

```python
def derivative_B(i,n,x): #Returns the derivative of the i-th Bernstein polynomial
    # of degree n evaluated in the point x.
    if n==0:
        return 0
    if i==0:
        return -n*standard_B(i,n-1,x)
    elif i==n:
        return n*standard_B(i-1,n-1,x)
    else:
        return n*(standard_B(i-1,n-1,x)-standard_B(i,n-1,x))
```

For the definition of $derivative\_B$ I used the formula:

$$ \frac{d}{dx}B_{i,n}(x) = n(B_{i-1,n-1}(x)-B_{i,n-1}(x) $$

5) Given the basis functions above, construct the Bernstein approximation of a continuous function, defined as:
$$B_n(f,x)=\sum_{i=0}^nB(i,n,x)f(i/n)$$
which is similar to what happens for the Lagrange interpolation. Use only the most efficient version of the above functions.

```python
def B_approx(n,f,x): #Returns the Bernstein approximation of degree n
    # of the function f evaluated in the point x.
    y = 0
    for i in range(n+1):
        y += standard_B(i,n,x)*f(i*1.0/n)
    return y

def Runge(x):
    return 1.0/(1.0 + (5.0-10.0*x)**2)

figure()
plot(x,B_approx(20,Runge,x),'b')
plot(x,Runge(x),'g')
```

I plotted the Bernstein approximation of degree $20$ of the Runge function to have a visual feedback.

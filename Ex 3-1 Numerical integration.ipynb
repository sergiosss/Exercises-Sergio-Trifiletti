1) Construct a interpolatory quadrature function, which takes as input a set of quadrature points and weights, a function, and returns the approximate integral computed using the given points and weights.

```python
def I(q, w, f):
    s = 0
    for i in range(len(q)):
        s += w[i]*f(q[i])
    return s
```

2) Construct a function that uses your scipy version of the lagrange basis functions to construct the quadrature weights w, given the quadrature points q and an integration interval (a,b). This function should compute the weights exploiting the fact that you can integrate numpy polynomials.

```python
class lagrange_basis: #A lagrange_basis object must be initialized with the set q
    # of interpolation points. Then it can be called using an index i in [0,n] to
    # return the i-th Lagrange polynomial as a polynomial object. It can also be
    # used with the method 'integral' with arguments i (index as before) and the
    # extremes of integration a and b.
    def __init__(self, q):
        self.q = q
        
    def __call__(self, i):
        myf = self.q*0
        myf[i] = 1
        return lagrange(self.q, myf)
    
    def integral(self, i, a, b):
        return self(i).integ()(b)-self(i).integ()(a)

class MyIntegral: #A MyIntegral object must be initialized with the set q of
    # interplation points, the extremes of integration as optional arguments
    # (if not given they are computed as the first and last interpolation points.
    # Note that this is necessary for quadrature formulas in which the interpolation
    # points do not include the extremes of integration, e.g. the mid-point formula),
    # and optionally the corresponding weights (if not given they will be computed).
    # Then it can be called using a function f as argument to compute the quadrature
    # formula given by the integration of the Lagrange polynomial of f on the nodes q.
    def compute_weigths(self):
        li = lagrange_basis(self.q)
        return np.array([li.integral(i, self.a, self.b) for i in range(len(self.q))])
    
    def __init__(self, q, a = None , b = None, w = None):
        self.q = q
        if a is None:
            self.a = self.q[0]
        else:
            self.a = a
        if b is None:
            self.b = self.q[-1]
        else:
            self.b = b
        if w is None:
            self.w = self.compute_weigths()
        else:
            self.w = w
        
    def __call__(self, f):
        return np.sum(self.w*f(self.q))
```

3) Construct an iterated quadrature formula, which takes weights and quadrature points constructed on the interval (0,1), a number of subdivisions N, and returns the quadrature points and weights (q_iter, w_iter) corresponding to the given quadrature formula rescaled on (i*1./(N-1), (i+1)*1./(N-1)) and repeated for each subinterval.

```python
class IteratedIntegral: #An IteratedIntegral object must be initialized with the
    # set q of interpolation points on the interval [0,1] (which will be rescaled
    # on each subinterval), the number N of subintervals, the extremes of integration
    # [a,b] (if not given will use [0,1]), and optionally the corresponding weights
    # (if not given they will be computed). Then it can be called using a function f
    # as argument to compute the iterated quadrature formula of f.
    def compute_weigths(self):
        li = lagrange_basis(self.q)
        return np.array([li.integral(i, 0.0, 1.0) for i in range(len(self.q))]) #The integral is between 0 and 1
    # because the weights are computed on the "original" set of interpolation points, which are given on [0,1]!
    
    def __init__(self, q, N, a = 0.0, b = 1.0, w = None):
        self.q = q
        self.N = N
        self.a = a
        self.b = b
        if w is None:
            self.w = self.compute_weigths()
        else:
            self.w = w
    
    def __call__(self, f):
        s = 0
        for i in range(self.N):
            w_iter = self.w / N
            c = self.a + (self.b-self.a)*i*1.0/(self.N)
            d = self.a + (self.b-self.a)*(i+1)*1.0/(self.N) #[c,d] is the i-th subinterval
            q_iter = np.array(self.q)*(d-c)+c #rescale q to the subinterval [c,d]
            Iqt = MyIntegral(q_iter, c, d, w_iter)
            s += Iqt(f)
        return s
```

4) Compare the results you obtain in integrating exp(-x) as well as the Runge counter example on the interval (0,1) using:

    Newton Cotes quadrature formula of order N (equispaced N+1 Lagrange interpolation points) on (0,1)
    Interpolatory quadrature formula on N+1 Chebishev points
    Iterated Trapez rule on N subdivisions (a function with N+1 distinct quadrature points).
    
5) Plot the errors you get when varying N.

```python
import numpy.polynomial.chebyshev as cheb

Ns = range(10,21)
```

```python
g = lambda x: np.exp(-x)
res = 1.0-np.exp(-1) #The exact result of the integral of g between 0 and 1
error1 = lambda N: abs(res - MyIntegral(np.linspace(0,1,N+1))(g))
error2 = lambda N: abs(res - MyIntegral(cheb.chebgauss(N+1)[0][::-1]*.5+.5)(g))
error3 = lambda N: abs(res - IteratedIntegral(np.linspace(0,1,2),N)(g))
plot(Ns, [error1(N) for N in Ns], label = 'Newton-Cotes')
plot(Ns, [error2(N) for N in Ns], label = 'Chebyshev')
plot(Ns, [error3(N) for N in Ns], label = 'Iterated trapez')
legend(loc=2)
xlabel('N')
ylabel('error')
title('$f(x)=e^{-x}$')
```

```python
Runge = lambda x: 1.0/(1.0+(5.0-10.0*x)**2)
res = np.arctan(5)/5.0 #The exact result of the integral of the Runge function between 0 and 1
error1 = lambda N: abs(res - MyIntegral(np.linspace(0,1,N+1))(Runge))
error2 = lambda N: abs(res - MyIntegral(cheb.chebgauss(N+1)[0][::-1]*.5+.5)(Runge))
error3 = lambda N: abs(res - IteratedIntegral(np.linspace(0,1,2),N)(Runge))
plot(Ns, [error1(N) for N in Ns], label = 'Newton-Cotes')
plot(Ns, [error2(N) for N in Ns], label = 'Chebyshev')
plot(Ns, [error3(N) for N in Ns], label = 'Iterated trapez')
legend(loc=2)
xlabel('N')
ylabel('error')
title('$f(x)=1/(1+(5-10x)^2)$')
```

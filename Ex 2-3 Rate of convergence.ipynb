3) Compute, for each possible combination, the L^infty norm of the interpolation error. Which routine gives you the best result?

```python
def adj(y): #Returns a list of 1-element lists, taking the elements from
    # the array y. It is necessary for the function PiecewisePolynomial.
    a = []
    for i in range(len(y)):
        l = []
        l.append(y[i])
        a.append(l)
    return a
```

```python
def L_infty(f, a=0, b=1): #Returns the L-infty norm of the function f on the
    x = np.linspace(a,b,1025)
    return np.max(abs(f(x)))
```

```python
x = np.linspace(0,1,1025)
fun = []
fun.append(lambda x: np.sin(2.0*np.pi*x))
fun.append(lambda x: 1.0/(1.0 + (5.0-10.0*x)**2))
fun.append(lambda x: abs(x-0.5))
f_name = ['$f(x)=sin(2\pi x)$', '$f(x)=1/(1+(5-10x)^2)$', '$f(x)=|x-0.5|$']
nodes = ['equispaced', 'Chebyshev']
meths = ['interp1d', 'BarycentricInterpolator', 'KroghInterpolator', 'PiecewisePolynomial', 'PchipInterpolator',
    'Akima1DInterpolator', 'UnivariateSpline', 'InterpolatedUnivariateSpline', 'lagrange']
```

```python
errors = np.ndarray(shape=(41,9,2,3))                
for N in range(10,51):
    qs = []
    qs.append(np.linspace(0,1,N))
    qs.append(cheb.chebgauss(N)[0][::-1]*.5+.5)
    for j in range(len(qs)):
        q = qs[j]
        for fn in range(len(fun)):
            f = fun[fn]
            app = []
            app.append(interp1d(q, f(q)))
            app.append(BarycentricInterpolator(q, f(q)))
            app.append(KroghInterpolator(q, f(q)))
            app.append(PiecewisePolynomial(list(q), adj(f(q))))
            app.append(PchipInterpolator(q, f(q)))
            app.append(Akima1DInterpolator(q, f(q)))
            app.append(UnivariateSpline(q, f(q)))
            app.append(InterpolatedUnivariateSpline(q, f(q)))
            app.append(lagrange(q, f(q)))
            for i in range(len(app)):
                interp = app[i]
                errf = lambda x: f(x)-interp(x)
                errors[N-10,i,j,fn] = L_infty(errf,q[0],q[-1])
#
```

Now we have the 4-dimensional array 'errors' containing the $L^\infty$ norm of the interpolation error for all the combinations of number of points, method, distribution of points, and function.

Let's plot these data to have a visual representation of which method gives the best result.

```python
for fn in range(3):
    for j in range(2):
        figure(figsize=(10,8))
        for i in range(9):
            subplot(3,3,i+1)
            plot(range(10,51),errors[:,i,j,fn])
            xlabel('N')
            title(meths[i])
        suptitle(f_name[fn]+', '+nodes[j]+' points')
        #tight_layout()
#
```

In general, the 5 methods interp1d, PiecwisePolynomial, PchipInterpolator, Akima1DInterpolator and InterpolatedUnivariateSpline achieve the best results, in particular InterpolatedUnivariateSpline produces slightly smaller errors than the others.

4) Compute the rate of convergence of the methods.

Let's compute the rates of convergence $r$ with the formula:

$$r = \frac{ln(\frac{e_2}{e_1})}{ln(\frac{h_2}{h_1})}$$

with $h(N)=$average distance between two successive interpolation points$=1/(N-1)$.

```python
short_meths = ['interp1d', 'BarycInter', 'KroghInter', 'PcwisePoly', 'PchipInter', 'Akima1DInt', 'UnivSpline',
    'IntUnivSpl', 'lagrange']
figure(figsize=(10,25))
for j in range(1):
    for i in range(9):
        for fn in range(3):
            xxs = []
            yys = []
            k = 0
            for h1 in range(10,50):
                for h2 in range(h1+1, 51):
                    xxs.append(k)
                    e2 = errors[h2-10,i,j,fn]
                    e1 = errors[h1-10,i,j,fn]
                    r = np.log(e2/e1)/np.log((h1-1.0)/(h2-1.0))
                    yys.append(r)
                    k += 1
            subplot(9,3,3*i+fn+1)
            plot(xxs,yys,'.')
            title(short_meths[i]+', '+f_name[fn])
    suptitle(nodes[j]+' points')
```
